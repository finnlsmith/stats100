```python
def grab_squad_list(url_input):
    stats = ["player"]
    res = requests.get(url_input)
        ## The next two lines get around the issue with comments breaking the parsing.
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res.text),'lxml')
    all_tables = soup.findAll("tbody")
    what_i_want = all_tables[0]

    rows_player = what_i_want.find_all('tr')
    return rows_player
```

### parse name and position for player

```python

def parse_name_and_group(player_row):
    array_final = []
    name = player_row.find("th",{"data-stat": "player"})

    name_encoded = name.text.strip().encode()
    decoded_name = name_encoded.decode('utf-8')

    pos = player_row.find("td",{"data-stat": "position"})
    pos_encoded = pos.text.strip().encode()
    pos_decoded = pos_encoded.decode('utf-8')

    natl = player_row.find("td",{"data-stat": "nationality"})
    natl_encoded = natl.text.strip().encode()
    natl_decoded = natl_encoded.decode('utf-8')
    

    array_final.append(decoded_name)
    array_final.append(pos_decoded)
    return decoded_name, pos_decoded, natl_decoded


```

```python
def create_player_table(input_player_rows_list):
  
    player_array = []
    pos_array = []
    natl_array = []
    #club stats 
    mins_array_club_5yr = []
    starts_array_club_5yr = []
    #league stats
    mins_array_league_5yr = []
    starts_array_league_5yr = []
    #top 5 league stats 
    mins_array_top5league_5yr = []
    starts_array_top5league_5yr = []
    #other league stats 
    mins_array_other_5yr = []
    starts_array_other_5yr = []

    youth_array = []
    is_new_array = []
    is_at_club_array = []

    ###NEW
    prev_season_mins_array = []
    num_seasons_array = []
    prev_season_starts_array = []
    #league experience adjusted mins, 5 seasons 
    mins_adjusted_array_5 = [] 
    mins_adjusted_array_1 = []

    for i in range(0, len(input_player_rows_list)):
        row_current = input_player_rows_list[i]
        name_and_pos = parse_name_and_group(row_current)
        
        player_array.append(name_and_pos[0])

        pos_array.append(name_and_pos[1])

        natl_array.append(name_and_pos[2])

        mins_array_club_5yr.append(0)
        starts_array_club_5yr.append(0)

        mins_array_league_5yr.append(0)
        starts_array_league_5yr.append(0)

        mins_array_top5league_5yr.append(0)
        starts_array_top5league_5yr.append(0)

        mins_array_other_5yr.append(0)
        starts_array_other_5yr.append(0)

        is_new_array.append(False)
        is_at_club_array.append(True)
        youth_array.append('No')

        ###NEW
        num_seasons_array.append(0)
        prev_season_mins_array.append(0)
        prev_season_starts_array.append(0)
        #NEW - experience adjusted minutes array append  
        mins_adjusted_array_5.append(0)
        mins_adjusted_array_1.append(0)


    player_df_5yr= pd. DataFrame ({"Player": player_array, 
                            "Position": pos_array, 
                            "Nationality": natl_array,
                            "Prev season Minutes, CLUB": mins_array_club_5yr,
                            "Prev season Starts, CLUB": starts_array_club_5yr,
                            "Prev season Minutes, LEAGUE": mins_array_league_5yr,
                            "Prev season Starts, LEAGUE": starts_array_league_5yr,
                            "Prev season Minutes, TOP 5 LEAGUES": mins_array_top5league_5yr,
                            "Prev season Starts, TOP 5 LEAGUES": starts_array_top5league_5yr,
                            "Prev season Minutes, OUTSIDE TOP 5 LEAGUES": mins_array_other_5yr,
                            "Prev season Starts, OUTSIDE TOP 5 LEAGUES": starts_array_other_5yr,

                            ###NEW
                            "Number of seasons played": num_seasons_array,
                            "Previous season minutes played": prev_season_mins_array,
                            'Previous season starts': prev_season_starts_array,
                            #experience adjusted mins - NEW
                            "Experience-adjusted minutes, last 5 seasons": mins_adjusted_array_5,
                            "Experience-adjusted minutes, previous season": mins_adjusted_array_1,
                            
                            "New arrival?": is_new_array,
                            "Is at club?":is_at_club_array,
                            'Youth player': youth_array
                            })
    if (player_df_5yr.at[1, 'Player'] == 'Robert Sánchez'):
        player_df_5yr.at[1, 'Player'] = 'Robert Sanchez'
    return player_df_5yr
```

```python
def populate_minsdata_at_club(player_df_5yr, club_name, club_league):
    #for i in range(0, 11):
    for i in range(0, len(player_df_5yr)):
        player_current = player_df_5yr.loc[i, 'Player']

        subset_player = massive_dataset[massive_dataset['player'] == player_current].reset_index().drop('index', axis=1)

        subset_player = subset_player[subset_player['nationality']== player_df_5yr.loc[i, 'Nationality']].reset_index().drop('index', axis=1)

        ###NEW
        num_seasons_played = len(subset_player)
        player_df_5yr.at[i, 'Number of seasons played'] = len(subset_player['Season'].unique())

        if(len(subset_player) >= 1):
            if(len(subset_player['birth_year'].unique()) > 1):
                
                if(len(subset_player[subset_player['team'] == f"{club_name}"]) != 0):
                    correct_birthday = subset_player[subset_player['team'] == f"{club_name}"].reset_index().drop('index', axis=1).at[0,'birth_year']
                    subset_player = subset_player[subset_player['birth_year'] == correct_birthday].reset_index().drop('index', axis=1)
                elif(player_df_5yr.loc[i, 'Position'] != 'GK'):
                    correct_pos = subset_player[subset_player['position'] != 'GK'].reset_index().drop('index', axis=1)
            if(subset_player.at[0, 'team'] != f"{club_name}"):  
                player_df_5yr.at[i,'New arrival?'] = True
            else:
                player_df_5yr.at[i,'New arrival?'] = False
            
            ###NEW
            prev_season_data = subset_player[subset_player['Season'] == 2022].reset_index().drop('index', axis=1)
            if(len(prev_season_data) == 0):
                player_df_5yr.at[i,'Previous season minutes played'] = 0
                player_df_5yr.at[i,'Previous season starts'] = 0
            else:
                player_df_5yr.at[i,'Previous season minutes played'] += prev_season_data['minutes'].sum()
                player_df_5yr.at[i,'Previous season starts'] += prev_season_data['games_starts'].sum()

                ##NEW - add to the 1 season adjusted here
                strength_numerator_1yr = strength_large[(strength_large['Year'] == prev_season_data.at[0, 'Season']) & (strength_large['League Name'] == prev_season_data.at[0, 'League'])].reset_index().at[0, 'Strength index']
                strength_denominator_1yr = strength_large[(strength_large['Year'] == prev_season_data.at[0, 'Season']) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
                strength_factor_1yr = strength_numerator_1yr / strength_denominator_1yr
                player_df_5yr.at[i, "Experience-adjusted minutes, previous season"] = (strength_factor_1yr * prev_season_data['minutes'].sum())

            for j in range (0, len(subset_player)):
                if (subset_player.at[j, 'team'] == f"{club_name}"):
                    player_df_5yr.at[i,'Prev season Minutes, CLUB'] += subset_player.at[j, 'minutes']
                    player_df_5yr.at[i,'Prev season Starts, CLUB'] += subset_player.at[j, 'games_starts']
                if (subset_player.at[j, 'League'] == f"{club_league}"):
                    player_df_5yr.at[i,'Prev season Minutes, LEAGUE'] += subset_player.at[j, 'minutes']
                    player_df_5yr.at[i,'Prev season Starts, LEAGUE'] += subset_player.at[j, 'games_starts']
                elif (subset_player.at[j, 'League'] in top5_leagues):
                    player_df_5yr.at[i,'Prev season Minutes, TOP 5 LEAGUES'] += subset_player.at[j, 'minutes']
                    player_df_5yr.at[i,'Prev season Starts, TOP 5 LEAGUES'] += subset_player.at[j, 'games_starts']
                else:
                    player_df_5yr.at[i,'Prev season Minutes, OUTSIDE TOP 5 LEAGUES'] += subset_player.at[j, 'minutes']
                    player_df_5yr.at[i,'Prev season Starts, OUTSIDE TOP 5 LEAGUES'] += subset_player.at[j, 'games_starts']

                ##NEW
                strength_numerator = strength_large[(strength_large['Year'] == subset_player.at[j, 'Season']) & (strength_large['League Name'] == subset_player.at[j, 'League'])].reset_index().at[0, 'Strength index']
                if(subset_player.at[j, 'Season'] == 2023):
                    strength_denominator = strength_large[(strength_large['Year'] == 2022) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
                else:
                    strength_denominator = strength_large[(strength_large['Year'] == subset_player.at[j, 'Season']) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
                strength_factor = strength_numerator / strength_denominator

                ## NEW - add to the 5 years here 
                player_df_5yr.at[i, "Experience-adjusted minutes, last 5 seasons"] += (strength_factor * subset_player.at[j, 'minutes'])  
                
        else:
            player_df_5yr.at[i,'New arrival?'] = False
            player_df_5yr.at[i,'Youth player'] = 'Yes'
        club_pt = subset_player[subset_player['team'] == f"{club_name}"]
        if((len(club_pt) == 1) & (club_pt['games_starts'].sum() <= 15)):
            player_df_5yr.at[i,'New arrival?'] = True

        ###NEW 
        player_df_5yr.at[i, "Experience-adjusted minutes, last 5 seasons"] = round(player_df_5yr.at[i, "Experience-adjusted minutes, last 5 seasons"])
        player_df_5yr.at[i, "Experience-adjusted minutes, previous season"] = round(player_df_5yr.at[i, "Experience-adjusted minutes, previous season"])
            
    return player_df_5yr

        


    

```

```python
#FOR THE SET OF PLAYERS WHO WERE AT CHELSEA IN PREV SEASON BUT LEFT, 

#for i in range(0, 1):
def populate_minsdata_departed(players_to_query_input, player_df_5yr_input, club_name, club_league):

    for i in range(0, len(players_to_query_input)):
        to_query_player = players_to_query_input[i]

        player_5yr_mins_data_subset = massive_dataset[massive_dataset['player']== to_query_player].reset_index().drop('index', axis=1)
        minscounter_club = 0
        stcounter_club = 0

        minscounter_league = 0
        stcounter_league = 0

        minscounter_t5league = 0
        stcounter_t5league = 0

        minscounter_other = 0
        stcounter_other = 0

        ##NEW - PREV SEASON MINS
        prevszn_mins = 0
        prevszn_starts = 0
        prevszn_mins_adjusted = 0

        prev_5_seasons_mins_adjusted = 0 



        ##NEW - NUM SEASONS PLAYED 
        num_seasons_played = len(player_5yr_mins_data_subset['Season'].unique())

        ###NEW - PREV SEASON MINS / STARTS
        player_prev_season_data = player_5yr_mins_data_subset[player_5yr_mins_data_subset['Season'] == 2022].reset_index().drop('index', axis=1)
        prevszn_mins += player_prev_season_data['minutes'].sum()
        prevszn_starts += player_prev_season_data['games_starts'].sum()

        ##NEW - calculate the 1 season experience-adjusted mins here
        strength_numerator_1yr = strength_large[(strength_large['Year'] == player_prev_season_data.at[0, 'Season']) & (strength_large['League Name'] == player_prev_season_data.at[0, 'League'])].reset_index().at[0, 'Strength index']
        strength_denominator_1yr = strength_large[(strength_large['Year'] == player_prev_season_data.at[0, 'Season']) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
        strength_factor_1yr = strength_numerator_1yr / strength_denominator_1yr
        prevszn_mins_adjusted = (strength_factor_1yr * player_prev_season_data['minutes'].sum())
        

        for j in range(0, len(player_5yr_mins_data_subset)):
            if(player_5yr_mins_data_subset.at[j, 'team'] == f"{club_name}"):
                minscounter_club += player_5yr_mins_data_subset.at[j, 'minutes']
                stcounter_club += player_5yr_mins_data_subset.at[j, 'games_starts']
            if(player_5yr_mins_data_subset.at[j, 'League'] == f"{club_league}"):
                minscounter_league += player_5yr_mins_data_subset.at[j, 'minutes']
                stcounter_league += player_5yr_mins_data_subset.at[j, 'games_starts']
            elif(player_5yr_mins_data_subset.at[j, 'League'] in top5_leagues):
                minscounter_t5league += player_5yr_mins_data_subset.at[j, 'minutes']
                stcounter_t5league += player_5yr_mins_data_subset.at[j, 'games_starts']
            else:
                minscounter_other += player_5yr_mins_data_subset.at[j, 'minutes']
                stcounter_other += player_5yr_mins_data_subset.at[j, 'games_starts']


            ##NEW - add to the 5 years here
            strength_numerator = strength_large[(strength_large['Year'] == player_5yr_mins_data_subset.at[j, 'Season']) & (strength_large['League Name'] == player_5yr_mins_data_subset.at[j, 'League'])].reset_index().at[0, 'Strength index']
            if(player_5yr_mins_data_subset.at[j, 'Season'] == 2023):
                strength_denominator = strength_large[(strength_large['Year'] == 2022) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
            else:
                strength_denominator = strength_large[(strength_large['Year'] == player_5yr_mins_data_subset.at[j, 'Season']) & (strength_large['League Name'] == club_league)].reset_index().at[0, 'Strength index']
            strength_factor = strength_numerator / strength_denominator

            prev_5_seasons_mins_adjusted += (strength_factor * player_5yr_mins_data_subset.at[j, 'minutes'])

        ##ROUND BEFORE ADDING
        prev_5_seasons_mins_adjusted = round(prev_5_seasons_mins_adjusted)
        new_row_5 = {
            'Player': player_5yr_mins_data_subset.at[0, 'player'],
            'Position':player_5yr_mins_data_subset['position'].unique()[0],
            'Nationality':player_5yr_mins_data_subset['nationality'].unique()[0],
            'Prev season Minutes, CLUB': minscounter_club,
            'Prev season Starts, CLUB': stcounter_club,
            'Prev season Minutes, LEAGUE': minscounter_league,
            'Prev season Starts, LEAGUE': stcounter_league,
            'Prev season Minutes, TOP 5 LEAGUES': minscounter_t5league,
            'Prev season Starts, TOP 5 LEAGUES': stcounter_t5league,
            'Prev season Minutes, OUTSIDE TOP 5 LEAGUES': minscounter_other,
            'Prev season Starts, OUTSIDE TOP 5 LEAGUES': stcounter_other,

            ##NEW METRICS ADDED
            'Number of seasons played': num_seasons_played,
            'Previous season minutes played': prevszn_mins,
            'Previous season starts': prevszn_starts,

            ##ADD HERE Experience-adjusted mins (5 szns and 1)
            'Experience-adjusted minutes, last 5 seasons': prev_5_seasons_mins_adjusted,
            'Experience-adjusted minutes, previous season': prevszn_mins_adjusted,
            'New arrival?': False,
            'Is at club?': False, 
            'Youth player': 'No'}

        player_df_5yr_input = pd.concat([player_df_5yr_input, pd.DataFrame([new_row_5])], ignore_index=True)

    return player_df_5yr_input
```

```python
def generate_strength_table(input_season):
    url_needed = f"https://www.teamform.com/en/league-ranking/world/{input_season}/q3"
    res_test = requests.get(url_needed)
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res_test.text),'lxml')
    all_tables = soup.findAll("tbody")
    what_i_want = all_tables[0]
    rows_leagues = what_i_want.find_all('tr')

    league_name_array = []
    country_array = []
    index_array = []
    year_array = []

    for row in rows_leagues:
        info = row.findAll("td")
        league_name = info[3].text.strip().encode().decode('utf-8')
        country_name = info[4].text.strip().encode().decode('utf-8')
        strength_index = info[5].text.strip().encode().decode('utf-8')

        league_name_array.append(league_name)
        country_array.append(country_name)
        index_array.append(strength_index)
        year_array.append(input_season)

    league_strength_DF = pd.DataFrame(
        {"League Name": league_name_array,
            "League Nation":country_array, 
            "Strength index": index_array, 
            "Year": year_array}
    )
    return league_strength_DF
       
```
